```python
# analyze and plot distribution of the unmodified weak labels vs all perturbed weak labels IN TRAINING SETS
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# COVIDNEWS
lab_stats_strat = pd.read_csv('../data/COVIDNEWS/descriptives/label_stats_strat.csv', index_col=0)
lab_stats_unif = pd.read_csv('../data/COVIDNEWS/descriptives/label_stats_unif.csv', index_col=0)
```

```python
lab_stats_strat
```

```python
lab_stats_unif
```

```python
df = lab_stats_unif.copy()
df.set_index('Label', inplace=True)

# Plot the counts for each distribution
df.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Label')
plt.ylabel('Count')
plt.yscale('log')
plt.title('Distribution of Counts for Different Labels')
plt.legend(title='Distribution Type')
plt.xticks(rotation=45)
plt.show()
```

```python
df = lab_stats_strat.copy()
df.set_index('Label', inplace=True)

# Plot the counts for each distribution
df.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Label')
plt.ylabel('Count')
plt.yscale('log')
plt.title('Distribution of Counts for Different Labels')
plt.legend(title='Distribution Type')
plt.xticks(rotation=45)
plt.show()
```
