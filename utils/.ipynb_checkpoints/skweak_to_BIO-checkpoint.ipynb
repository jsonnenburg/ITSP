```python
import pandas as pd
import itertools
import os
import ast
```

```python
data_train = pd.read_csv("../data/COVIDNEWS/data_perturbed/raw/data_train_perturbed_strat.csv", index_col=0)
```

```python
data_train
```

```python
# function needs to work on tokenized labels --> apply to all starts_with("labels_tok") cols
```

```python
relev_cols = [col for col in data_train if col.startswith('labels_tok')]
```

```python
for column in relev_cols:
    data_train[column] = data_train[column].astype(str)
    data_train[column] = data_train[column].apply(lambda x: ast.literal_eval(x))
```

```python
data_train.labels_tok[5]
```

```python
def generate_bio_labels(entity_labels):
    bio_labels = []
    for label in entity_labels:
        if label == 'O':
            bio_labels.append(label)
        else:
            if bio_labels and bio_labels[-1] == label:
                bio_labels.append('I-' + label.lower())
            else:
                bio_labels.append('B-' + label.lower())
    return bio_labels
```

```python
bio_labels = generate_bio_labels(data_train.labels_tok[5])
```
